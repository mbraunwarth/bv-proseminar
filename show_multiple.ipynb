```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from skimage import (io, data, filters, util, restoration)
```

```python
image = data.coins()
noisy = util.random_noise(image)
gauss = filters.gaussian(noisy)
bilat = restoration.denoise_bilateral(noisy, multichannel=False)
```

```python
fig, axes = plt.subplots(2, 2)
ax = axes.ravel()

ax[0].imshow(image, cmap=cm.gray)
ax[0].set_title("Original image")

ax[1].imshow(noisy, cmap=cm.gray)
ax[1].set_title("Noisy")

ax[2].imshow(gauss, cmap=cm.gray)
ax[2].set_title("Gaussian filtered")

ax[3].imshow(bilat, cmap=cm.gray)
ax[3].set_title("Bilateral filtered")

fig.tight_layout()
```
