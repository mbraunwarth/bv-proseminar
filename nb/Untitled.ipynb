```python
from skimage import (io, data, restoration, util)
import numpy as np
```

```python
img = data.coffee()
io.imshow(img)
io.show()
```

```python
noisy = util.random_noise(img)
io.imshow(noisy)
io.show()
```

```python
filtered = restoration.denoise_bilateral(noisy, multichannel=True)
io.imshow(filtered)
io.show()
```

```python
#img.shape
#img[:,:,0] -> red channel
img_max = lambda x: max(img[:,:,x].flatten())
print('red mean: {0}'.format(np.mean(img[:,:,0]/img_max(0))))
print('green mean: {0}'.format(np.mean(img[:,:,1]/img_max(1))))
print('blue mean: {0}'.format(np.mean(img[:,:,2]/img_max(2))))
```

```python
print('red mean: {0}'.format(np.mean(filtered[:,:,0])))
print('green mean: {0}'.format(np.mean(filtered[:,:,1])))
print('blue mean: {0}'.format(np.mean(filtered[:,:,2])))
```

```python
print("-- Errors --")
print("Red\t", np.mean(filtered[:,:,0]) - np.mean(img[:,:,0]/img_max(0)))
print("Green\t", np.mean(filtered[:,:,1]) - np.mean(img[:,:,1]/img_max(1)))
print("Blue\t", np.mean(filtered[:,:,2]) - np.mean(img[:,:,2]/img_max(2)))
```
